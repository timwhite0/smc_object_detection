```python
import matplotlib.pyplot as plt
import seaborn as sns
import torch

device="cpu"
```

### Load results

```python
true_counts = torch.load("../images/true_counts.pt").to(device)
true_fluors = torch.load("../images/true_fluors.pt").to(device)
true_locs = torch.load("../images/true_locs.pt").to(device)
true_axes = torch.load("../images/true_axes.pt").to(device)
true_angles = torch.load("../images/true_angles.pt").to(device)
true_total_intensities = torch.load("../images/true_total_intensities.pt").to(device)
images = torch.load("../images/images.pt").to(device)
num_images = images.shape[0]
max_objects = true_fluors.shape[1]

posterior_mean_count = torch.load("posterior_mean_count.pt").to(device)
posterior_mean_total_fluor = torch.load("posterior_mean_total_fluor.pt").to(device)
reconstructed_image = torch.load("reconstructed_image.pt").to(device)
runtime = torch.load("runtime.pt").to(device)
num_iters = torch.load("num_iters.pt").to(device)

watershed_estimated_counts = torch.load("watershed_estimated_counts.pt").to(device)
```

### SMC summary

```python
print(f"RMSE (cell count) across {images.shape[0]} images:",
      ((posterior_mean_count - true_counts)**2).mean().sqrt().item())
print(f"MAE (cell count) across {images.shape[0]} images:",
      ((posterior_mean_count - true_counts).abs()).mean().item())
print(f"RMSE (total fluorescence) across {images.shape[0]} images:",
      ((posterior_mean_total_fluor - true_fluors.sum(1))**2).mean().sqrt().item())
print(f"MAE (total fluorescence) across {images.shape[0]} images:",
      ((posterior_mean_total_fluor - true_fluors.sum(1)).abs()).mean().item())
print(f"correct number of cells detected in {(posterior_mean_count.round() == true_counts).sum()} of the {images.shape[0]} images (accuracy = {((posterior_mean_count.round() == true_counts).sum()/images.shape[0])})")
print(f"runtime: minimum = {runtime.min().int()}, mean = {runtime.mean().int()}, median = {runtime.median().int()}, maximum = {runtime.max().int()}")
print(f"number of iterations: minimum = {num_iters.min().int()}, mean = {num_iters.mean().int()}, median = {num_iters.median().int()}, maximum = {num_iters.max().int()}")
```

```python
smc_mean_count_by_count = torch.zeros(max_objects+1)
smc_rmse_by_count = torch.zeros(max_objects+1)
smc_mae_by_count = torch.zeros(max_objects+1)
smc_rmse_total_fluor_by_count = torch.zeros(max_objects+1)
smc_mae_total_fluor_by_count = torch.zeros(max_objects+1)
smc_num_correct_by_count = torch.zeros(max_objects+1)
smc_prop_correct_by_count = torch.zeros(max_objects+1)

for num in range(max_objects+1):
    print(f"true count = {num}")
    
    smc_mean_count_by_count[num] = posterior_mean_count[true_counts==num].mean()
    print(f"estimated count for images where count = {num}:", smc_mean_count_by_count[num].item())
    
    smc_rmse_by_count[num] = ((posterior_mean_count[true_counts==num] - true_counts[true_counts==num])**2).mean().sqrt()
    print(f"RMSE (count) across {(true_counts==num).sum()} images:", (smc_rmse_by_count[num].item()))
    
    smc_mae_by_count[num] = ((posterior_mean_count[true_counts==num] - true_counts[true_counts==num]).abs()).mean()
    print(f"MAE (count) across {(true_counts==num).sum()} images:", (smc_mae_by_count[num].item()))
    
    smc_rmse_total_fluor_by_count[num] = ((posterior_mean_total_fluor[true_counts==num] - true_fluors[true_counts==num].sum(1))**2).mean().sqrt()
    print(f"RMSE (total fluor) across {(true_counts==num).sum()} images:", (smc_rmse_total_fluor_by_count[num].item()))
    
    smc_mae_total_fluor_by_count[num] = ((posterior_mean_total_fluor[true_counts==num] - true_fluors[true_counts==num].sum(1)).abs()).mean()
    print(f"MAE (total fluor) across {(true_counts==num).sum()} images:", (smc_mae_total_fluor_by_count[num].item()))
    
    smc_num_correct_by_count[num] = (posterior_mean_count[true_counts==num].round() == true_counts[true_counts==num]).sum()
    smc_prop_correct_by_count[num] = smc_num_correct_by_count[num]/(true_counts==num).sum()
    print(f"correct count in {smc_num_correct_by_count[num].int()} of the {(true_counts==num).sum()} images (accuracy = {smc_prop_correct_by_count[num]})")
    
    print(f"runtime: minimum = {runtime[true_counts==num].min().int()}, median = {runtime[true_counts==num].median().int()}, maximum = {runtime[true_counts==num].max().int()}")
    print(f"number of iterations: minimum = {num_iters[true_counts==num].min().int()}, median = {num_iters[true_counts==num].median().int()}, maximum = {num_iters[true_counts==num].max().int()}\n")
```

```python
print(f"RMSE (cell count) across {images.shape[0]} images:",
      ((watershed_estimated_counts - true_counts)**2).mean().sqrt().item())
print(f"MAE (cell count) across {images.shape[0]} images:",
      ((watershed_estimated_counts - true_counts).abs()).mean().item())
print(f"correct number of cells detected in {(watershed_estimated_counts.round() == true_counts).sum()} of the {images.shape[0]} images (accuracy = {((watershed_estimated_counts.round() == true_counts).sum()/images.shape[0])})")
```

### Figure 2: Reconstructed images

```python
torch.manual_seed(10)

fontsize = 16
cmap = 'cividis'

example_image_indexes = [torch.arange(num_images)[true_counts==1][torch.randint(low=0, high=(true_counts==1).sum(), size=(1,))].item(),
                         torch.arange(num_images)[true_counts==2][torch.randint(low=0, high=(true_counts==2).sum(), size=(1,))].item(),
                         torch.arange(num_images)[true_counts==3][torch.randint(low=0, high=(true_counts==3).sum(), size=(1,))].item(),
                         torch.arange(num_images)[true_counts==4][torch.randint(low=0, high=(true_counts==4).sum(), size=(1,))].item()]
fig2_counts = true_counts[example_image_indexes]
fig2_original = images[example_image_indexes]
fig2_smc = reconstructed_image[example_image_indexes]

fig2, ax = plt.subplots(nrows = 4, ncols = 2, sharex = True, sharey = True, figsize = (6,12))
fig2.tight_layout(h_pad = -0.5, w_pad = -0.5)

for i in range(len(example_image_indexes)):
    f2_orig = ax[i,0].imshow(fig2_original[i].cpu(), cmap = cmap)
    _ = ax[i,0].set_ylabel(f"count = {fig2_counts[i].int()}", fontsize = fontsize)
    # _ = ax[i,0].xaxis.set_label_position("top")
    _ = ax[i,0].set_xticks([])
    _ = ax[i,0].set_yticks([])
    cbar = plt.colorbar(f2_orig, ax = ax[i,0], shrink = 0.6)
    cbar.ax.tick_params(labelsize = fontsize)

    f2_smc = ax[i,1].imshow(fig2_smc[i].cpu(), cmap = cmap)
    _ = ax[i,1].set_xticks([])
    _ = ax[i,1].set_yticks([])
    cbar = plt.colorbar(f2_orig, ax = ax[i,1], shrink = 0.6)
    cbar.ax.tick_params(labelsize = fontsize)

_ = ax[0,0].set_xlabel("Original", loc = "center", fontsize = fontsize)
_ = ax[0,0].xaxis.set_label_position("top")
_ = ax[0,1].set_xlabel("SMC", loc = "center", fontsize = fontsize)
_ = ax[0,1].xaxis.set_label_position("top")

fig2.subplots_adjust(wspace=0.2,hspace=-0.2)
fig2.savefig('figures/figure2.png', bbox_inches = "tight", dpi=500)
```

### Figure 3: Accuracy, calibration, and mean absolute error of cell counts

```python
# Bootstrap CIs for classification accuracy
torch.manual_seed(0)

num_boot_samples = 10000
smc_bounds_prop_correct_by_count = torch.zeros(max_objects+1, 2)
watershed_bounds_prop_correct_by_count = torch.zeros(max_objects+1, 2)

### SMC
for num in range(max_objects+1):
    boot_orig = torch.stack((posterior_mean_count[true_counts==num].round(), true_counts[true_counts==num]), dim = 1)
    boot_prop_correct = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_prop_correct[j] = (boot_sample[:,0] == boot_sample[:,1]).sum()/boot_orig.size(0)
    
    smc_bounds_prop_correct_by_count[num] = boot_prop_correct.quantile(torch.tensor((0.05, 0.95),device=device))

### Watershed
for num in range(max_objects+1):
    boot_orig = torch.stack((watershed_estimated_counts[true_counts==num].round(), true_counts[true_counts==num]), dim = 1)
    boot_prop_correct = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_prop_correct[j] = (boot_sample[:,0] == boot_sample[:,1]).sum()/boot_orig.size(0)
    
    watershed_bounds_prop_correct_by_count[num] = boot_prop_correct.quantile(torch.tensor((0.05, 0.95),device=device))
```

```python
# Bootstrap CIs for mean estimated count
torch.manual_seed(0)

num_boot_samples = 10000
smc_bounds_count_by_count = torch.zeros(max_objects+1, 2)
watershed_bounds_count_by_count = torch.zeros(max_objects+1, 2)

### SMC
for num in range(max_objects+1):
    boot_orig = posterior_mean_count[true_counts==num].round()
    boot_mean_count = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_mean_count[j] = boot_sample.mean()
    
    smc_bounds_count_by_count[num] = boot_mean_count.quantile(torch.tensor((0.05, 0.95),device=device))

### Watershed
for num in range(max_objects+1):
    boot_orig = watershed_estimated_counts[true_counts==num].round()
    boot_mean_count = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_mean_count[j] = boot_sample.mean()
    
    watershed_bounds_count_by_count[num] = boot_mean_count.quantile(torch.tensor((0.05, 0.95),device=device))
```

```python
# Bootstrap CIs for mean absolute error
torch.manual_seed(0)

num_boot_samples = 10000
smc_bounds_mae_by_count = torch.zeros(max_objects+1, 2)
watershed_bounds_mae_by_count = torch.zeros(max_objects+1, 2)

### SMC
for num in range(max_objects+1):
    boot_orig = 1.*(posterior_mean_count[true_counts==num] - true_counts[true_counts==num]).abs()
    boot_mae = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_mae[j] = boot_sample.mean()
    
    smc_bounds_mae_by_count[num] = boot_mae.quantile(torch.tensor((0.05, 0.95),device=device))

### Watershed
for num in range(max_objects+1):
    boot_orig = 1.*(watershed_estimated_counts[true_counts==num] - true_counts[true_counts==num]).abs()
    boot_mae = torch.zeros(num_boot_samples)
    
    for j in range(num_boot_samples):
        boot_sample = boot_orig[(torch.multinomial(torch.arange(boot_orig.size(0)).double(), num_samples = boot_orig.size(0), replacement = True))]
        boot_mae[j] = boot_sample.mean()
    
    watershed_bounds_mae_by_count[num] = boot_mae.quantile(torch.tensor((0.05, 0.95),device=device))
```

```python
dot_size = 75
linewidth = 2
text_size = 16
smc_color = "saddlebrown"
watershed_color = "teal"

range_max_objects = torch.arange(max_objects+1, device="cpu")

fig3, (fig3a,fig3b,fig3c) = plt.subplots(nrows = 3, ncols = 1, figsize = (6,12))
fig3.tight_layout(w_pad = 5)

_ = fig3a.plot(range_max_objects, smc_prop_correct_by_count.cpu(), label = "SMC", color = smc_color, linewidth = linewidth)
_ = fig3a.scatter(range_max_objects, smc_prop_correct_by_count.cpu(), color = smc_color, s = dot_size)
_ = fig3a.fill_between(range_max_objects, smc_bounds_prop_correct_by_count[:,0].cpu(), smc_bounds_prop_correct_by_count[:,1].cpu(), color = smc_color, alpha = 0.1)
_ = fig3a.plot(range_max_objects, watershed_prop_correct_by_count.cpu(), label = "Watershed", color = watershed_color, linewidth = linewidth)
_ = fig3a.scatter(range_max_objects, watershed_prop_correct_by_count.cpu(), color = watershed_color, s = dot_size)
_ = fig3a.fill_between(range_max_objects, watershed_bounds_prop_correct_by_count[:,0].cpu(), watershed_bounds_prop_correct_by_count[:,1].cpu(), color = watershed_color, alpha = 0.1)
_ = fig3a.legend(ncol=2, fancybox=True, loc="upper center", bbox_to_anchor=(0.5,1.2), fontsize = text_size)
_ = fig3a.tick_params(axis = "both", labelsize = text_size)
_ = fig3a.set_xlabel("True cell count", fontsize = text_size)
_ = fig3a.set_ylabel("Classification accuracy", fontsize = text_size)

_ = fig3b.plot(range_max_objects, smc_mean_count_by_count.cpu(), color = smc_color, linewidth = linewidth)
_ = fig3b.scatter(range_max_objects, smc_mean_count_by_count.cpu(), color = smc_color, s = dot_size)
_ = fig3b.fill_between(range_max_objects, smc_bounds_count_by_count[:,0].cpu(), smc_bounds_count_by_count[:,1].cpu(), color = smc_color, alpha = 0.1)
_ = fig3b.plot(range_max_objects, watershed_mean_count_by_count.cpu(), color = watershed_color, linewidth = linewidth)
_ = fig3b.scatter(range_max_objects, watershed_mean_count_by_count.cpu(), color = watershed_color, s = dot_size)
_ = fig3b.fill_between(range_max_objects, watershed_bounds_count_by_count[:,0].cpu(), watershed_bounds_count_by_count[:,1].cpu(), color = watershed_color, alpha = 0.1)
_ = fig3b.plot(range_max_objects, range_max_objects, ":", color="black", linewidth = linewidth)
_ = fig3b.tick_params(axis = "both", labelsize = text_size)
_ = fig3b.set_xlabel("True cell count", fontsize = text_size)
_ = fig3b.set_ylabel("Mean estimated cell count", fontsize = text_size)

_ = fig3c.plot(range_max_objects, smc_mae_by_count.cpu(), color = smc_color, linewidth = linewidth)
_ = fig3c.scatter(range_max_objects, smc_mae_by_count.cpu(), color = smc_color, s = dot_size)
_ = fig3c.fill_between(range_max_objects, smc_bounds_mae_by_count[:,0].cpu(), smc_bounds_mae_by_count[:,1].cpu(), color = smc_color, alpha = 0.1)
_ = fig3c.plot(range_max_objects, watershed_mae_by_count.cpu(), color = watershed_color, linewidth = linewidth)
_ = fig3c.scatter(range_max_objects, watershed_mae_by_count.cpu(), color = watershed_color, s = dot_size)
_ = fig3c.fill_between(range_max_objects, watershed_bounds_mae_by_count[:,0].cpu(), watershed_bounds_mae_by_count[:,1].cpu(), color = watershed_color, alpha = 0.1)
_ = fig3c.tick_params(axis = "both", labelsize = text_size)
_ = fig3c.set_xlabel("True cell count", fontsize = text_size)
_ = fig3c.set_ylabel("Mean absolute error", fontsize = text_size)

fig3.subplots_adjust(hspace=0.5)
fig3.savefig('figures/figure3.png', bbox_inches = "tight", dpi = 500)
```
